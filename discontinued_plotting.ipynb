Showing how to make a dicontinued plot. Sort of!

```python
# imports, give names, avoid importing *, no one likes *, it's just sloppy...
import numpy as np
from matplotlib import pyplot as plt
```

```python
delta = 1e-2  # let's say this is our resolution
```

```python
def fa(x):
    return x**3

def fb(x):
    return x**2

# make a function defined for 0 < x < 10, with a discontinuity with the size 2 delta inbetween
def f(x):
    split = 3        # define the inconsistency
    x = np.array(x)  # make sure we have an array
    out = x * 0      # just make a copy to initialize
    
    # we could have set these with if's and but's, but array notation is sweeter!
    out[:] = None    # set all to None
    maska = x < (split - delta)
    out[maska] = fa(x[maska])  # set lower values
    maskb = x > (split + delta)
    out[maskb] = fb(x[maskb])  # set higher values
    out[x<=0] = None
    out[x>=10] = None
    
    return out
```

```python
plt.figure(dpi=200)
x = np.arange(-3, 5, step=delta)
y = f(x)
plt.plot(x, y)
```
